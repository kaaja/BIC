# Maximization of continues single variable function with biological algo

$f(x)  = - x^4 + 2 x^3 + 2x^2 -x$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
seaborn.set(style="white", context="notebook", font_scale=1.5,
            rc={"axes.grid": True, "legend.frameon": False,
"lines.markeredgewidth": 1.4, "lines.markersize": 10})
```

```python
# Data generating function
start = -2
stop = 3
numberOfPoints = 500
x = np.linspace(start, stop, numberOfPoints)
f = lambda x: -x**4 + 2*x**3 + 2*x**2 - x

# Initial population
np.random.seed(seed=1)
newPopulation = np.random.uniform(-2, 3, 8)

# Plotting initial population
fig, ax = plt.subplots()
ax.plot(x, f(x), label='f', linewidth=4.2)
ax.plot(newPopulation, f(newPopulation), 'o')
ax.set_title('Initial population')
plt.show()

numberOfParentCandidates = 4
iterations = 25

for iteration in range(iterations):
    # Parent candidates
    parantCandidatesIndices = np.random.choice(len(newPopulation), numberOfParentCandidates, replace=False)
    print('parantCandidatesIndices', parantCandidatesIndices)

    # Choose two best parent candidates as parents
    parentCandidates = newPopulation[parantCandidatesIndices]
    print('parentCandidates', parentCandidates)
    fitnessParentCandidates = f(parentCandidates)
    print('fitnessParentCandidates', fitnessParentCandidates)
    numberOfParents = 2
    bestParantCandidatesIndices = np.argpartition(fitnessParentCandidates, numberOfParents)[numberOfParents:]
    print('bestParantCandidatesIndices', bestParantCandidatesIndices)
    
    # Create children with random weighted average of parents + some mutation randomness
    inheritanceShareParent1 = np.random.uniform(0,1)
    crossoverChild1 = inheritanceShareParent1*parentCandidates[bestParantCandidatesIndices[0]] + (1 - inheritanceShareParent1)*parentCandidates[bestParantCandidatesIndices[1]]
    mutationRate = np.random.uniform(-2,2)
    child1 = crossoverChild1 + mutationRate*np.max(np.asarray(abs(parentCandidates[bestParantCandidatesIndices[0]]), abs(parentCandidates[bestParantCandidatesIndices[1]])))
    # Avoid searching outside domain
    if child1 < -2 or child1 >3:
        child1 = parentCandidates[bestParantCandidatesIndices[0]]

    inheritanceShareParent1 = np.random.uniform(0,1)
    crossoverChild2 = inheritanceShareParent1*parentCandidates[bestParantCandidatesIndices[0]] + (1-inheritanceShareParent1)*parentCandidates[bestParantCandidatesIndices[1]] 
    mutationRate = np.random.uniform(-2,2)
    child2 = crossoverChild2 + mutationRate*np.max(np.asarray(abs(parentCandidates[bestParantCandidatesIndices[0]]), abs(parentCandidates[bestParantCandidatesIndices[1]])))
    if child2 < -2 or child2 >3:
        child2 = parentCandidates[bestParantCandidatesIndices[1]]
    
    # Remove 2 worst individuals from total population
    populationWith2many = (np.append(newPopulation, [child1, child2]))
    print('populationWith2many', populationWith2many, '\n')
    fitnessPopulationWith2many = f(populationWith2many)
    print('fitnessPopulationWith2many',fitnessPopulationWith2many, '\n' )
    newPopulationIndices = np.argpartition(fitnessPopulationWith2many, len(newPopulation-2))[2:]
    print('newPopulationIndices', newPopulationIndices, '\n')
    newPopulation = populationWith2many[newPopulationIndices]
    print('newPopulation', newPopulation, '\n')
    
    # For plotting
    maxFitness = np.max(f(newPopulation))
    maxFitnessIndex = np.argmax(f(newPopulation))
    xMax = newPopulation[maxFitnessIndex]
    
    if iteration % 5 == 0:
        fig, ax = plt.subplots()
        ax.plot(x, f(x), label='f', linewidth=4.2)
        ax.plot(newPopulation, f(newPopulation), 'o', label='new')
        ax.set_title('Iteration %d  x* = %.1f  f(x*) = %.1f' %(iteration+1, xMax, maxFitness))
        plt.show()
```
